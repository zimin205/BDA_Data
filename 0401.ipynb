```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
x = np.array([2, 4, 6, 8])
```

```python
y = np.array([81, 92, 91, 97])
```

```python
plt.scatter(x,y)
plt.show()
```

```python
a= 0
b= 0
```

```python
lr = 0.03
epochs = 2001
```

```python
n = len(x)
```

```python
for i in range(epochs):
  y_pred = a*x+b
  error = y - y_pred

  a_diff = (2/n) * sum (-x * (error))
  b_diff = (2/n) * sum(-(error))

  a = a - lr*a_diff
  b = b - lr*b_diff
  if i % 100 == 0:
    print("epoch=%.f, 기울기=%.04f, 절편=%.04f"%(i, a, b))

```

```python
y_pred = a * x + b
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred, 'r')
plt.show()
```
