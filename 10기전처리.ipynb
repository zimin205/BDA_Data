```python
file_path = '/content/drive/MyDrive/BDA 6기 운영진/6기 데이터팀/김지민/10기 출석표.xlsx'


출석표 = pd.read_excel(file_path, dtype={'phone_number': str})
```

```python
import pandas as pd

# "분반" 컬럼을 사용하여 데이터 정리
출석표_강의 = 출석표[['식별자', '분반']].dropna().drop_duplicates()
출석표_강의 = 출석표_강의.groupby('식별자')['분반'].apply(list).reset_index()

# class1_update, class2_update, class3_update, class4_update 컬럼 생성
for idx, row in 출석표_강의.iterrows():
    분반명_list = row['분반']

    # class1_update부터 class4_update까지 업데이트
    for i in range(4):
        if i < len(분반명_list):
            출석표_강의.at[idx, f'class{i+1}'] = 분반명_list[i]
        else:
            출석표_강의.at[idx, f'class{i+1}'] = None

# 숫자만 변환하는 함수 정의
def convert_to_int_or_keep(value):
    if isinstance(value, str) and value.isdigit():  # 문자열이 숫자로만 되어 있으면 변환
        return int(value)
    return value  # 변환 불가능하면 그대로 반환

# class1~class4 컬럼에 적용
for i in range(1, 5):
    column_name = f'class{i}'
    출석표_강의[column_name] = 출석표_강의[column_name].apply(convert_to_int_or_keep)

print(출석표_강의)  # 결과 확인
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report

import pandas as pd

tdf = pd.read_excel('/content/drive/MyDrive/BDA 6기 운영진/6기 데이터팀/김지민/8_9_df.xlsx')
X = tdf['대학명(학사)']
y = tdf['대학명(학사)_최종']

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier

# TF-IDF로 텍스트를 숫자로 변환
vectorizer = TfidfVectorizer(max_features=5000, min_df=5, max_df=0.7)
X_vectorized = vectorizer.fit_transform(X)
#train, test 분리
X_train, X_test, y_train, y_test = train_test_split(X_vectorized,y, test_size = 0.3, random_state = 111)

from imblearn.combine import SMOTETomek
from imblearn.over_sampling import SMOTE

smote_tomek = SMOTETomek(random_state=111, smote=SMOTE(sampling_strategy='minority', k_neighbors=1))
X_resampled, y_resampled = smote_tomek.fit_resample(X_train, y_train)

from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

model = SVC(probability=True)

# 하이퍼파라미터 탐색 범위 설정
param_dist = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto'],
    'class_weight': ['balanced']
}

# RandomizedSearchCV로 최적의 하이퍼파라미터 탐색
random_search = RandomizedSearchCV(model, param_distributions=param_dist, n_iter=10, cv=3, random_state=111)
random_search.fit(X_train, y_train)

# 최적의 하이퍼파라미터로 모델 재훈련
best_model = random_search.best_estimator_
best_model.fit(X_train, y_train)

# 예측
y_pred = best_model.predict(X_test)
y_pred_prob = best_model.predict_proba(X_test)[:, 1]

# 성능 평가
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')  # 'weighted'로 설정
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-Score: {f1}")
```
