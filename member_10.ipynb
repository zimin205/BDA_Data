```python
member_gu_df=  pd.read_excel('/content/drive/MyDrive/BDA 6기 운영진/6기 데이터팀/김지민/member_구.xlsx')
before_id_df=  pd.read_excel('/content/drive/MyDrive/BDA 6기 운영진/6기 데이터팀/김지민/before_id.xlsx')
member_9_df = pd.read_excel('/content/drive/MyDrive/BDA 6기 운영진/6기 데이터팀/김지민/member.xlsx')
```

```python
# NaN을 빈 문자열로 바꾸고, 소수점 없애고, 문자열로 고정
member_gu_df['phone_number'] = (
    member_gu_df['phone_number']
    .fillna("")                      # NaN -> 빈 문자열
    .apply(lambda x: str(int(x)) if isinstance(x, float) else str(x))  # float -> int -> str
    .str.zfill(10)                  # 앞자리 0 보존
)
member_9_df['phone_number'] = (
    member_9_df['phone_number']
    .fillna("")                      # NaN -> 빈 문자열
    .apply(lambda x: str(int(x)) if isinstance(x, float) else str(x))  # float -> int -> str
    .str.zfill(10)                  # 앞자리 0 보존
)
```

```python
# 🧼 전화번호 정리 & 식별자 생성 함수
def clean_phone_and_create_id(df):
    df['phone_number'] = (
        df['phone_number']
        .fillna("")
        .apply(lambda x: str(int(x)) if isinstance(x, float) else str(x))
        .str.zfill(10)
    )
    df['식별자'] = df['member_name'] + '_' + df['phone_number'].str[-4:-2]
    return df

# 모든 DF 정제 및 식별자 생성
before_id_df = clean_phone_and_create_id(before_id_df)
member_gu_df = clean_phone_and_create_id(member_gu_df)
member_9_df = clean_phone_and_create_id(member_9_df)

# 📌 3~8기: previous_class_8 → previous_class_3 역순 순회
for gen in range(8, 2, -1):
    prev_class_col = f'previous_class_{gen}'

    # 해당 기수 수강한 사람만 추출
    target = before_id_df[before_id_df[prev_class_col].notna() & (before_id_df[prev_class_col] != '해당없음')]

    # 해당 기수 데이터
    prev_members = member_gu_df[member_gu_df['generation'] == gen][['식별자', 'member_id']].copy()
    prev_members['member_id'] = prev_members['member_id'].astype(str)

    match_dict = dict(zip(prev_members['식별자'], prev_members['member_id']))

    def assign_before_id(row):
        if pd.notna(row.get('before_id')):
            return row['before_id']
        return match_dict.get(row['식별자'], None)

    before_id_df['before_id'] = before_id_df.apply(assign_before_id, axis=1)

# 🔁 9기 처리 (member.xlsx 기반)
prev_members_9 = member_9_df[['식별자', 'member_id']].copy()
prev_members_9['member_id'] = prev_members_9['member_id'].astype(str)
match_dict_9 = dict(zip(prev_members_9['식별자'], prev_members_9['member_id']))

def assign_before_id_9(row):
    if pd.notna(row.get('before_id')):
        return row['before_id']
    return match_dict_9.get(row['식별자'], None)

before_id_df['before_id'] = before_id_df.apply(assign_before_id_9, axis=1)

# 💾 저장
before_id_df.to_excel("before_id_final.xlsx", index=False)

```
